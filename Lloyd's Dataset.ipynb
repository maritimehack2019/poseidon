## How to find n closest ports to any given coordinates

```python
# function to return n closest ports
def closest(lat, long, n):
    dist = [mpu.haversine_distance(x, (lat, long)) for x in list(zip(ports['LATITUDE DECIMAL'], ports['LONGITUDE DECIMAL']))]
    closest = sorted(enumerate(dist), key = lambda x: x[1])[:n]
    return ports.iloc[[i[0] for i in closest]]
```

```python
closest(1, 2, 10)
```

```python
# putting the n closest ports on a map
import folium
```

```python
def mapper(lat, long, n):
    result = folium.Map(location=[lat, long], zoom_start=10)
    results = closest(lat, long, n)
    coords = results[['LATITUDE DECIMAL', 'LONGITUDE DECIMAL']].values.tolist()
    folium.Marker(location=[lat, long],
                  popup='YOU ARE HERE',
                  icon=folium.Icon(color='red')).add_to(result)
    for i in range(0, n):
        point = results.iloc[i]
        folium.Marker(location=coords[i],
              popup=point['PLACE NAME'],
              icon=folium.Icon(color='blue')).add_to(result)
    return result
```

```python
# function to open the closest port in Google Maps
def linkclosest(lat, long):
    result = closest(lat, long, 1)
    url = 'http://www.google.com/maps/place/'+ str(result['LATITUDE DECIMAL'].values[0])+','+str(result['LONGITUDE DECIMAL'].values[0])
    chrome_path = 'open -a /Applications/Google\ Chrome.app %s'
    webbrowser.get(chrome_path).open(url) 
    return url
    
```
